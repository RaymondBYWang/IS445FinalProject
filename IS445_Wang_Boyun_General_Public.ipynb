```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
import seaborn as sns
import json
import re
import random
from functools import reduce
from ast import literal_eval
```

```python
df = pd.read_csv('/content/IDOR_2009_Illinois_Individual_Income_Tax_Web_Returns_Retirement_by_County.csv')
df[:100]
```

```python
#Chart 1： Scatter Plot for Number of Return
plt.scatter(df.index, df['returns'], alpha=0.6)
plt.title('Scatter Plot of Number of Returns')
plt.xlabel('County Index')
plt.ylabel('Number of Returns')
plt.grid(True)
plt.show()
```

```python
#Chart 2： Scatter Plot for Total Tax
plt.scatter(df.index, df['total_tax'], alpha=0.6)
plt.title('Scatter Plot of Total Tax')
plt.xlabel('County Index')
plt.ylabel('Total Tax ($)')
plt.grid(True)
plt.show()
```

```python
#Chart 3： Bar Plot for Number of Return
plt.bar(df['county'], df['returns'])
plt.title('Number of Returns per County')
plt.xlabel('County')
plt.ylabel('Returns')
plt.xticks(rotation=90)
plt.show()
```

```python
#Chart 4： Bar Plot for Total Tax
plt.bar(df['county'], df['total_tax'])
plt.title('Total Tax per County')
plt.xlabel('County')
plt.ylabel('Total Tax ($)')
plt.xticks(rotation=90)
plt.show()
```
